```python
# The goal is to identify linear correlations

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corr = df.corr()
df

# Positive correlations: Positive correlations can be seen in the two scatter
# plots I gave below.  Basically, when a value increases on one scale, another
# increases with it.  This makes the values in direct positive correlation - when 
# one variable increases, so does the other.  For example, the higher control a 
# player has over a ball, the greater skill they have in dribbling.

# Negative correlations: Negative correlations are different - when an x variable
# increases, the y value decreases.  This infers that the data points at the 0,0 point
# of the graph is higher than the rest of the data points.  Although it's a little
# harder to find an example here, It can be seen in the bar graph showcasing Age correlating
# with International Reputation. If we were measuring players from ages 30-45, we would have
# an example of a negative correlation.  

# Which columns I would use as inputs and targets: 
# I would use Age as an input/ independent variable/ x-axis value because it
# has an easily tracable limit - it's categorical.  What I mean by this is that there  
# is a very small number of 'groups' the data can be placed in, seeing as the numerical
# range for age is so small.  I would use the rest of the bulk data/ measurable data as the 
# targets (GKDiving, GKKicking, Release Clause) because they can act as continuous variables.  
# Basically, instead of just having a small range of numbers associated with measuring the data, 
# (for Age, it would be 19-45) there is a larger bulk of data that can be associated with all of
# the individual players. 
```

```python
sns.heatmap(corr)
```

```python
plt.title('Heatmap Featuring Positive and Negative Correlations')
plt.show()
```

```python
plt.scatter(x=df.BallControl, y=df.Dribbling)
plt.title('Ball Control Correlation to Dribbling')
plt.xlabel('Ball Control')
plt.ylabel('Dribbling')

plt.show()

# Some insights we can conglean from this visualization: 
# We can see that the more a player has control over a ball, the 
# better they are at dribbling the ball.  Although this is probably
# obvious, it's good to see the correlation between the two skills - 
# and how they directly correlate with one another. 
```

```python
plt.scatter(x=df.ShotPower, y=df.LongShots)
plt.title('ShotPower Correlation to LongShots')
plt.xlabel('Shot Power')
plt.ylabel('Long Shots')

plt.show()

# Some insights we can conglean from this visualization: 
# Similar to the graph above, and not too surprising, the shot power a player has pretty much directly 
# correlates with the long shots.  This means that the greater shot power
# a player has, the longer the shot will be.  
```

```python
# Also, if you change 'bar' to 'plot' you get a cool design -not useful, just cool... kinda like me
plt.bar(df.Age, df['International Reputation'])
plt.title('FIFA 2019 Player Reputation by Age')
plt.ylabel('International Reputation')
plt.xlabel('Age')

plt.show()

# Some insights we can conglean from this visualization: 
# We can see that it creates a bell graph - so
# the reputation of players raises then falls with age.  
# We learn a little about what age players gain international 
# popularity (around 25-37 years) and how younger and older 
# players are less popular - basically, reputation correlates
# with age in a way that shows which players are more well-known
# across the world
```
